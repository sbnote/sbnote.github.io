```python
import tensorflow as tf
```

### Broadcastable?
- Match from Last dim
  - (1) If current dim=1, expand to same 
  - (2) If either has no dim, insert one dim and expand to same
  - otherwise, NOT broadcastable

### Broadcasting

```python
x = tf.random.normal([4, 32, 32, 3])
(x+tf.random.normal([4,1,1,1]))  # (1)
```

```python
(x+tf.random.normal([3])).shape  # (2) & (1)
```

```python
(x+tf.random.normal([32,32,1])).shape  # (2) & (1)
```

### tf.broadcast_to

```python
b=tf.broadcast_to(tf.random.normal([4,1,1,1]), [4,32,32,3])
b.shape
```

理解：与Broadcasting部分的区别，broadcast_to是将tf.random.normal([4,1,1,1])扩展填充成[4,32,32,3]的形式； 而Broadcasting部分，是加法运算需要tf.random.normal([4,1,1,1])也拥有与tf.random.normal([4, 32, 32, 3])一样的形式，故其相当于隐式broadcast_to成[4,32,32,3]

### Broadcast VS Tile

```python
a = tf.ones([3,4])
a2 = tf.expand_dims(a, axis=0)  # TensorShape([1, 3, 4])
a2 = tf.tile(a2, [2,1,1])  # 第1维复制2份，第2，第3维复制1份（即不变）
a2.shape  # TensorShape([2, 3, 4])
```

理解：同是扩展填充，Tile是复制填充，内存占用大；而Broadcast非实际复制，内存占用更小。
